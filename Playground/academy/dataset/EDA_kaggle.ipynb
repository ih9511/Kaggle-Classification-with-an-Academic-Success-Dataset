## 2. Train Data

```python
# Categorical data columns
categorical_columns = [
    'Marital status',
    'Application mode',
    'Course',
    'Daytime/evening attendance',
    'Previous qualification',
    'Nacionality',
    'Mother\'s qualification',
    'Father\'s qualification',
    'Mother\'s occupation',
    'Father\'s occupation',
    'Displaced',
    'Educational special needs',
    'Debtor',
    'Tuition fees up to date',
    'Gender',
    'Scholarship holder',
    'International',
]
```

```python
# numerical data columns
numerical_columns = [
    'Application order',
    'Previous qualification (grade)',
    'Age at enrollment',
    'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)',
    'Unemployment rate',
    'Inflation rate',
    'GDP',
    # 'Target'
]
```

```python
for col in categorical_columns:
    train_data[col] = train_data[col].astype('category')
```

```python
for col in numerical_columns:
    train_data[col] = pd.to_numeric(train_data[col])
```

### 2.1 Numerical Columns

```python
# 수치형 변수 분석
numeric_columns = train_data.select_dtypes(include = [np.number]).columns
```

```python
train_data[numeric_columns].hist(bins = 30, figsize = (15, 10))
plt.show()
```

```python
corr_matrix = train_data[numeric_columns].corr()
sns.heatmap(corr_matrix, annot = False)
plt.title('Correlation Matrix')
plt.show()
```

### 2.2 Categorical Columns

```python
# 범주형 변수 분석
categorical_columns = train_data.select_dtypes(include = [object, 'category']).columns

for col in categorical_columns:
    plt.figure(figsize = (10, 5))
    sns.countplot(x = col, data = train_data)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# 타겟 변수와의 관계
for col in categorical_columns:
    if col != 'Target':
        plt.figure(figsize = (10, 5))
        sns.countplot(x = col, hue = 'Target', data = train_data)
        plt.title(f'{col} vs Target')
        plt.show()
```

```python
# target 변수별 특성 비교
for col in numeric_columns:
    plt.figure(figsize = (10, 5))
    sns.boxplot(x = 'Target', y = col, data = train_data)
    plt.title(f'{col} vs Target')
    plt.show()
```

## 3. 이상치 분석

```python
for col in numeric_columns:
    plt.figure(figsize = (10, 5))
    sns.boxplot(x = train_data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
